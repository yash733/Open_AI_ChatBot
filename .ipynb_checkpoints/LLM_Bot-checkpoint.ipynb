```python
import openai

openai.api_key = '#ENTER_YOUR_KEY'
```

```python
import streamlit as st

def generate_response(user_msg):
    # Example: Use the Excel data to generate a response
    if "nutrition in" in user_msg.lower():
        item = user_msg.lower().split("nutrition in")[-1].strip()
        nutrition_info = get_nutrition_info(item)
        if nutrition_info:
            return f"Nutrition information for {item}: {nutrition_info}"
        else:
            return f"Sorry, I don't have information on {item}."
    else:
        # Default response using GPT model
        return get_gpt_response(user_msg)

def get_nutrition_info(item):
    # Look up the nutrition information in the Excel data
    item_data = data[data['Name'].str.lower() == item.lower()]
    if not item_data.empty:
        return item_data.iloc[0].to_dict()
    return None

def get_gpt_response(user_msg):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=user_msg,
        max_tokens=150
    )
    return response.choices[0].text.strip()

# Streamlit app
st.title("Food Nutrition Information Chatbot")
st.write("Ask me anything about food nutrition!")

# Input box for user to type their question
user_input = st.text_input("You: ", "")

# When user hits enter, generate the response
if user_input:
    response = generate_response(user_input)
    st.text_area("Bot:", value=response, height=200, max_chars=None)
```
